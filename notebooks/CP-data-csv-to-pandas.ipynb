Transform results from CP stored in csv files to pandas databases
1. Generating joined_data dataframe
2. Generating individual channel dataframes

1. Generating joined_data dataframe

```python
import pandas

path_to_csv1 = "/mnt/efs/aimbl_2025/student_data/S-DD/LD088-LD089-LD090-LD093-Nucleoli2-Nuclei.csv"
path_to_csv2 = "/mnt/efs/aimbl_2025/student_data/S-DD/LD088-LD089-LD090-LD093-Nucleoli2-A488objPerNucleus.csv"
path_to_csv3 = "/mnt/efs/aimbl_2025/student_data/S-DD/LD088-LD089-LD090-LD093-Nucleoli2-A568objPerNucleus.csv"

data1 = pandas.read_csv(path_to_csv1, header=0, index_col=[0, 1])
data2 = pandas.read_csv(path_to_csv2, header=0, index_col=[0, 1])
data3 = pandas.read_csv(path_to_csv3, header=0, index_col=[0, 1])
data1clean = data1.drop(columns=['Metadata_Cells', 'Metadata_Channel1', 'Metadata_Channel2', 'Metadata_Channel3', 'Metadata_Experiment', 'Metadata_Field', 'Metadata_FileLocation', 'Metadata_Frame', 'Metadata_Microscope', 'Metadata_Obj', 'Metadata_Series', 'Metadata_Treatment.1', 'Metadata_Zstacks', 'FileName_Image', 'PathName_Image'])
data2clean = data2.drop(columns=['Metadata_Cells', 'Metadata_Channel1', 'Metadata_Channel2', 'Metadata_Channel3', 'Metadata_Experiment', 'Metadata_Field', 'Metadata_FileLocation', 'Metadata_Frame', 'Metadata_Microscope', 'Metadata_Obj', 'Metadata_Series', 'Metadata_Treatment', 'Metadata_Treatment.1', 'Metadata_Zstacks', 'FileName_Image', 'PathName_Image'])
data3clean = data3.drop(columns=['Metadata_Cells', 'Metadata_Channel1', 'Metadata_Channel2', 'Metadata_Channel3', 'Metadata_Experiment', 'Metadata_Field', 'Metadata_FileLocation', 'Metadata_Frame', 'Metadata_Microscope', 'Metadata_Obj', 'Metadata_Series', 'Metadata_Treatment', 'Metadata_Treatment.1', 'Metadata_Zstacks', 'FileName_Image', 'PathName_Image'])
data1clean_DAPI = data1clean.add_prefix('DAPI_')
data2clean_A488 = data2clean.add_prefix('A488_')
data3clean_A568 = data3clean.add_prefix('A568_')
joined_data = pandas.concat([data1clean_DAPI, data2clean_A488, data3clean_A568], axis=1)
joined_data = joined_data.fillna(value=0)
joined_data = joined_data.rename(columns={'DAPI_Metadata_Treatment' : 'Metadata_Treatment'})

print(f"Header DAPI data header: {data1clean_DAPI.head()}")
print(f"Header A488 data: {data2clean_A488.head()}")
print(f"Header A568 data: {data3clean_A568.head()}")
print(f"Header joined data: {joined_data.head()}")
```
